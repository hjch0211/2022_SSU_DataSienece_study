```python
import seaborn as sns
import pandas as pd

titanic = sns.load_dataset("titanic")

#titanic.isnull().sum() #결측값 확인
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
#value_counts()로 값들을 set처럼 가져옴
titanic['embarked'] = titanic['embarked'].fillna('S')
titanic['embark_town'] = titanic['embark_town'].fillna('Southampton')
titanic['deck'] = titanic['deck'].fillna('C')
titanic.info()
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(1, 2, figsize=(10, 5))
titanic['survived'][titanic['sex']=='male'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%', ax=ax[0], shadow=True)
titanic['survived'][titanic['sex']=='female'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%', ax=ax[1], shadow=True)
ax[0].set_title('male')
ax[1].set_title('female')
```

```python
sns.countplot('pclass', hue = 'survived',data=titanic)
plt.show()
```

```python
titanic_corr = titanic.corr(method='pearson') #피어슨 상관 계수 구하기
#titanic['survived'].corr(titanic['adult_male']) 특정 변수 지정 가능
sns.pairplot(titanic, hue='survived') #왜 에러 뜨지
plt.show()
```

```python
sns.catplot(x='pclass',y='survived',hue='sex',data=titanic,kind='point') #hue로 비교하는 것같음
plt.show()
```

```python
sns.heatmap(titanic.corr(), annot=True) #annot이 true면 값까지
```
