```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

plt.rc('font', family='NanumGothic')
print(plt.rcParams['font.family'])


soongsil= pd.read_csv('soongsil.csv', encoding = 'cp949', engine = 'python')
soongsil_bar=soongsil['indsLclsNm'].value_counts()
soongsildf=pd.DataFrame(soongsil_bar)
soongsildf1=soongsildf.reset_index()
soongsildf1.rename(columns={"index": "종류","indsLclsNm":"합계"}, inplace=True)

plt.figure(figsize=(12,8))  #size 조절
plt.xticks(rotation = 0)  #x축 조절

sns.barplot(
data = soongsildf1.sort_values(by="종류", ascending=False),
x = "종류",
y = "합계"
)
plt.show()
```
