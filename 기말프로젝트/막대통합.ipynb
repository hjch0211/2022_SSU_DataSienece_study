```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

plt.rc('font', family='NanumGothic')
print(plt.rcParams['font.family'])
soongsil= pd.read_csv('soongsil.csv', encoding = 'cp949', engine = 'python')
chungang= pd.read_csv('chungang.csv', encoding = 'cp949', engine = 'python')
yeonsae= pd.read_csv('yeonsae.csv', encoding = 'cp949', engine = 'python')
seoul= pd.read_csv('seoul.csv', encoding = 'cp949', engine = 'python')
donguk= pd.read_csv('donguk.csv', encoding = 'cp949', engine = 'python')


def getUnivBar(csv):
    csv_bar=csv['indsLclsNm'].value_counts()
    csvdf=pd.DataFrame(csv_bar)
    csvdf1=csvdf.reset_index()
    csvdf1.rename(columns={"index": "종류","indsLclsNm":"합계"}, inplace=True)
    plt.figure(figsize=(12,8))  #size 조절
    plt.xticks(rotation = 0)  #x축 조절
    sns.barplot(
    data = csvdf1.sort_values(by="종류", ascending=False),
    x = "종류",
    y = "합계"
    )
    plt.show()
    
getUnivBar(soongsil)
ss=soongsil['indsLclsNm'].value_counts()
cc=chungang['indsLclsNm'].value_counts()
yy=yeonsae['indsLclsNm'].value_counts()
se=seoul['indsLclsNm'].value_counts()
dd=donguk['indsLclsNm'].value_counts()
```
